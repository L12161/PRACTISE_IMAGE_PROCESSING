```python
# we try to generate two images with the same histogram. this is possible when the image pixels are random permutated

print("I'm rooting for Lakers, so that they lose")
```

```python
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
from PIL import Image
```

```python
image_nilam = cv2.imread("/content/photo_2021-05-31_02-31-47.jpg",1)

cv2_imshow(image_nilam)
```

```python
import numpy as np
random_img = np.random.permutation(image_nilam);
```

```python
cv2_imshow(random_img)
```

```python
color = ('b','g','r')

for i,col in enumerate(color):
  print(i)
  HISTOGRAM=cv2.calcHist([image_nilam],[i],None,[256],[0,256])
  plt.plot(HISTOGRAM,color=col)
  plt.xlim([0,256])

plt.show()
```

```python
color = ('b','g','r')

for i,col in enumerate(color):
  HISTOGRAM=cv2.calcHist([random_img],[i],None,[256],[0,256])
  plt.plot(HISTOGRAM,color=col)
  plt.xlim([0,256])

plt.show()
```
