```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_nilam = cv2.imread("/content/71851b51ec2559587c95fede8afb6c24.jpg",1)
```

```python
plt.imshow(image_nilam)
plt.show()
```

```python
color = ('b','g','r')

for i,col in enumerate(color):
  HISTOGRAM=cv2.calcHist([image_nilam],[i],None,[256],[0,256])
  plt.plot(HISTOGRAM,color=col)
  plt.xlim([0,256])

plt.show()
```

```python
# now we get the image dimension
H,W,_ = image_nilam.shape
```

```python
# FIRST METHOD OF IMAGE INVERSION 
# subtracting all the pixels

for i in range(0,H-1):
  for j in range(0,W-1):

    #pixel under consideration
    pixel = image_nilam[i,j]

    # considering the red, blue and green colours for a certain pixel

    pixel[0]=255-pixel[0]
    pixel[1]=255-pixel[1]
    pixel[2]=255-pixel[2]

    # insert the pixel back to the image
    image_nilam[i,j]=pixel



```

```python
plt.imshow(image_nilam)
plt.show()
```

```python
color = ('b','g','r')

for i,col in enumerate(color):
  HISTOGRAM=cv2.calcHist([image_nilam],[i],None,[256],[0,256])
  plt.plot(HISTOGRAM,color=col)
  plt.xlim([0,256])

plt.show()
```

```python
# alternative approaches can be taken to convert the images to negetive 
# just  subtract the image from 1

image_abar_nilam = cv2.imread("/content/71851b51ec2559587c95fede8afb6c24.jpg",1)
```
